Type Casting + Comments

```python
x=str(3)
y=int(3)
z=float(3)
print(x,y,z)
"""
the above statement would gave the output of x y z
"""
print(type(x)) #shows the type of x like its datatype
print(type(y))
print(type(z))
```

Case sensitive

```python
A="John" #the capital A stores the string
a=10  #the small a stores the integer
print(A)
print(a)
```

#String datatype


String Slicing

```python
b="Sheheryar Ahmad"
print(b[3:9]) #it would just cut of the string from the index that is given to it but would not show the last one
print(b[0:10])
```

Lower and Upper Case

```python
a="Sheheryar Ahmad"
print(a.lower())
print(a.upper())
b=" Ahmad "
print(b.strip()) #it removes the white spaces from the start and last of strings
```

Concatenation

```python
a="Sheheryar"
b="Ahmad"
print(a+" "+b) #the + sign connect(concat) two or more strings
c=a+"  "+b
print(c)
```
