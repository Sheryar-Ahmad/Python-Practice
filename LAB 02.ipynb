Module 5: Electricity Bill System

```python
unit = int(input("Enter the unit: "))
bill = 0

if unit <= 100:
    # First 100 units at Rs. 5
    bill = unit * 5
elif unit <= 200:
    # First 100 at Rs. 5, remaining units at Rs. 8
    bill = (100 * 5) + ((unit - 100) * 8)
else:
    # First 100 at Rs. 5, next 100 at Rs. 8, above 200 at Rs. 10
    bill = (100 * 5) + (100 * 8) + ((unit - 200) * 10)

print(f"Your total bill is Rs. {bill}")
```

Module 8: Operator-Based Calculator


```python
# Get input from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, %, **): ")

# Perform calculation based on the operator
if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/":
    # Using if-else here to prevent division by zero
    result = num1 / num2 if num2 != 0 else "Error (Div by 0)"
elif op == "%":
    result = num1 % num2
elif op == "**":
    result = num1 ** num2
else:
    result = "Invalid Operator"

print(f"Result: {result}")
```
