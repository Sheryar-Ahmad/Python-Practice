TASK 7 Palindrome Check

```python
word=str(input("Enter the Word "))
if(word==word[::-1]):
    print("It is a Palindrome")
else:
    print("It is not a Palindrome")
```

Task 8 Character Count

```python
vowel=0
consonants=0

sentence=str(input("Enter the Sentence or Word"))
for i in range(len(sentence)):
    if sentence[i]in('a','e','i','o','u'):
      vowel=vowel+1
    else:
      consonants=consonants+1

print(vowel)
print(consonants)
```

TASK 11 Frequency of each charactr in string

```python
text=str(input("Enter the Text"))
freq={}
for char in text:
 freq[char]=freq.get(char,0)+1

print(freq)
```
