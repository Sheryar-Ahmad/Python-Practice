Task 1: Creating Arrays (15 Minutes)

Task 2: Special Arrays (15 Minutes)
Create the following arrays using NumPy functions:
a) Array of zeros with size 5.
b) Array of ones with size 8.
c) Array containing numbers from 1 to 20.
d) Array containing even numbers from 2 to 20.
e) Array with 10 equally spaced values between 0 and 1.

```python
import numpy as np

# Task a: Array of zeros with size 5
zeros_array = np.zeros(5)
print("a) Array of zeros (size 5):")
print(zeros_array)
print()

# Task b: Array of ones with size 8
ones_array = np.ones(8)
print("b) Array of ones (size 8):")
print(ones_array)
print()

# Task c: Array containing numbers from 1 to 20
range_array = np.arange(1, 21)
print("c) Numbers from 1 to 20:")
print(range_array)
print()

# Task d: Array containing even numbers from 2 to 20
even_array = np.arange(2, 21, 2)
print("d) Even numbers from 2 to 20:")
print(even_array)
print()

# Task e: Array with 10 equally spaced values between 0 and 1
linspace_array = np.linspace(0, 1, 10)
print("e) 10 equally spaced values between 0 and 1:")
print(linspace_array)
```

Task 3: Array Indexing and Slicing (20 Minutes)
Given the array:
[language=Python] arr = np.array([5,10,15,20,25,30,35,40])
Perform the following:
a) Print the first element.
b) Print the last element.
c) Print elements from index 2 to 5.
d) Print the first four elements.
e) Print all elements except the first two.


```python
import numpy as np

# Given array
arr = np.array([5, 10, 15, 20, 25, 30, 35, 40])

print("Original array:", arr)
print()

# Task a: Print the first element
print("a) First element:", arr[0])

# Task b: Print the last element
print("b) Last element:", arr[-1])

# Task c: Print elements from index 2 to 5
print("c) Elements from index 2 to 5:", arr[2:6])

# Task d: Print the first four elements
print("d) First four elements:", arr[:4])

# Task e: Print all elements except the first two
print("e) All elements except first two:", arr[2:])
```

Task 4: Mathematical Operations (20 Minutes)
Create:
[language=Python] A = np.array([1,2,3,4,5]) B = np.array([10,20,30,40,50])
Perform:
a) Addition
b) Subtraction
c) Multiplication
d) Division
e) Square each element of array A

```python
import numpy as np
# Create the arrays
A = np.array([1, 2, 3, 4, 5])
B = np.array([10, 20, 30, 40, 50])

print("Array A:", A)
print("Array B:", B)
print()

# Task a: Addition
addition = A + B
print("a) Addition (A + B):", addition)

# Task b: Subtraction
subtraction = A - B
print("b) Subtraction (A - B):", subtraction)

# Task c: Multiplication
multiplication = A * B
print("c) Multiplication (A * B):", multiplication)

# Task d: Division
division = A / B
print("d) Division (A / B):", division)

# Task e: Square each element of array A
squared = A ** 2
print("e) Square of each element in A:", squared)
```

Task 5: Two-Dimensional Arrays (20 Minutes)
Create the matrix:


2 4 6
8 10 12
14 16 18


Then:
a) Print the shape of the matrix.
b) Print the element in row 2, column 3.
c) Print the first row.
d) Print the second column.
e) Calculate the sum of all elements.

```python
import numpy as np
# Create the matrix
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])

print("Matrix:")
print(matrix)
print()

# Task a: Print the shape of the matrix
print("a) Shape of the matrix:", matrix.shape)

# Task b: Print the element in row 2, column 3
# Note: Python uses 0-based indexing, so row 2 is index 1, column 3 is index 2
print("b) Element at row 2, column 3:", matrix[1, 2])

# Task c: Print the first row
print("c) First row:", matrix[0, :])

# Task d: Print the second column
print("d) Second column:", matrix[:, 1])

# Task e: Calculate the sum of all elements
print("e) Sum of all elements:", np.sum(matrix))
```

Task 6: Statistical Functions (15 Minutes)
Given:
[language=Python] marks = np.array([78,85,92,67,88,95,73,81])
Find:
a) Maximum value
b) Minimum value
c) Mean
d) Median
e) Standard deviation

```python
import numpy as np

# Given array
marks = np.array([78, 85, 92, 67, 88, 95, 73, 81])

print("Marks array:", marks)
print()

# Task a: Maximum value
print("a) Maximum value:", np.max(marks))

# Task b: Minimum value
print("b) Minimum value:", np.min(marks))

# Task c: Mean
print("c) Mean:", np.mean(marks))

# Task d: Median
print("d) Median:", np.median(marks))

# Task e: Standard deviation
print("e) Standard deviation:", np.std(marks))
```

Challenge Task (15 Minutes)
Create a NumPy array containing the marks of 10 students.
Write a program to:
1. Calculate the average marks.
2. Find the highest and lowest marks.
3. Display all marks greater than the average

```python
import numpy as np

# Create array with marks of 10 students
marks = np.array([78, 85, 92, 67, 88, 95, 73, 81, 90, 76])

print("Marks of 10 students:", marks)
print()

# 1. Calculate the average marks
average = np.mean(marks)
print(f"1. Average marks: {average:.2f}")

# 2. Find the highest and lowest marks
highest = np.max(marks)
lowest = np.min(marks)
print(f"2. Highest marks: {highest}")
print(f"   Lowest marks: {lowest}")
print()

# 3. Display all marks greater than the average
marks_above_avg = marks[marks > average]
print(f"3. Marks greater than average ({average:.2f}):")
print(f"   {marks_above_avg}")
print(f"   Count: {len(marks_above_avg)} students")
```
