# **List in the python**
#list is changable,
#list is ordered,
#in list multiple items in one variable,
#it can allow the duplicate values,


```python
marks = [10,20,30,40,50]
print("\n Original Marks:",marks) # we can also show the elements of the list without the loop

print("\n ----Created the List that contains the names of the student-----")
name= ["Sheryar","Owais","Hammad","Tamjeed","Tanveer","Abdullah","Ahsan","Musa","Hamdan","Osama"]
print("\n Names of the Students are: ",name)

print("\n ----To show the element using the index-----")
# To show the element using the index
print("\n First Name: ",name[0])

print("\n ----Showing the names using Slicing-----")
#Showing the names using Slicing
print("\n Middle Names :", name[4:6])


print("\n ----Reversing the List Elements-----")
#Reversing the List Elements
name.reverse() # it can be reversed by both the ways like using the built in fucntion and also the slicing
print("\n Reversed List: ",name)
#print("Reversed List: ",name[::-1])  #[start:stop:step]


print("\n ----Adding a new Student name to the List-----")
#Adding a new Student name to the List
name.append("Daniyal")  #append key word would add the new student name in the end of the list
print("\n New List 1: ",name)



print("\n ----Inserting the student name at a specific index-----")
#Inserting the student name at a specific index
name.insert(3,"Tabish") # it will insert the element at the index 3 and move all the other elements by increment 1 so and no data would be lost
print("\n New List 2: ",name)


# Min Max and Average on the List having Numbers
print("\n -----USING THE MAX MIN AND THE AVERAGE FUNCTIONS----")
max_marks=max(marks)
print("\n Max Element is: ", max_marks)

min_marks=min(marks)
print("\n Min Element is: ", min_marks)

avg_marks=sum(marks)/len(marks)
print("\n Average Marks is: ", avg_marks)


print("\n----Checking the type of the list----")
print(type(name))

print("\n----Using the For Loop to print the Elements of the List----")
for x in name:  # for loop in the python
  print(x)
```

#**TUPLES**
#Tuples are used to store multiple items in a single variable
#A tuple is a collection which is ordered and unchangeable and also allow duplicate values.
#Tuples are written with round brackets.

```python
cities=("Islamabad","Lahore","Karachi","Peshawar","Quetta") # creating the tuple
print(cities)

#printing the first city
print("\n First City: ",cities[0])

#printing the last city
print("\n Last City: ",cities[-1])

#Middle city
print("\n Middle City: ",cities[-3])


#Printing all the cities one by one
for x in cities:
  print("\n One by One Elements: ",x)



#Printing the Number of the Elements of the Tuple
print("\n Length of the Tuple is ",len(cities))

#Checking the City that it exist in the tuple or not
for x in cities:
  if(x == "Lahore"):
    print(x,"Is present in the tuple" )


#converting the tuple to list and then adding a new city and then again converting it back

print("\n-----Converting the Tuple to List and then Adding a new city to it and then again converting it back to the Tuple-----")
cities=list(cities)
cities.append("Nowshera")
cities=tuple(cities)
print("\n New Tuple: ",cities)


print("\n------Using the count Keyword to find how much time any duplicated variable is present in the tuple------")
num=(1,2,3,2,4,2,6,7,8)
print("The Duplicated variable is present: ",num.count(2))

print("\n------Using the index Keyword to find on which index the duplicated value is present------")
for x in num:
  if(x == 2):
    print(x,"Is present at index: ",num.index(x))
#print("The Duplicated variable is present: ",num.index(2))
```

#**Sets**
#Sets are used to store multiple items in a single variable.
#A set is a collection which is unordered, unchangeable*, and unindexed.

```python
Ai_Student={"Shery","Ahmad","Osama","Ali","Taimur","mohaimin"}
Python_Student={"osama","Ahmad","Hassan","mohaimin"}
print("\nAi Student Name:",Ai_Student) # here the output would be shown as randomly
print("\nPython Student Name: ",Python_Student) # it would also print the student names

common=Ai_Student.intersection(Python_Student) # it would show the common names that are present in both the sets
print("\nCommon Student Name: ",common)

union=Ai_Student.union(Python_Student) # it would show all the elements present in the both sets
print("\nUnion Student Name: ",union)

difference=Ai_Student.difference(Python_Student) #it would remove all the elements
print("\nDifference Student Name: ",difference)





```
